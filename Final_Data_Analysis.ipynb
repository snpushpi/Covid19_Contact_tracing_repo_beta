```
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from cycler import cycler
plt.style.use('ggplot')
%matplotlib inline
```

```
main_intersection = [0,5,10,15,20,25,30,35,40,45,50,75,80,85]
l2 = len(main_intersection)
```

```
def TPR(tp, fn):
    frac = tp/(tp+fn)
    return 100*frac
```

```
def FPR(fp, tn):
    frac = fp/(fp+tn)
    return 100*frac
```

```
def FNR(fn, tp):
    frac = fn/(fn+tp)
    return 100*frac
```

```
def TNR(tn, fp):
    frac = tn/(tn+fp)
    return 100*frac
```

Final No Fake Path contact tracing same hash with run number=100

```
tp_fp_list=[(8,342),(7,352),(5,355),(5,355),(8,317),(7,338),(5,328),(9,340),(8,343),(6,299),(8,347),(11,301)]
epsilon_list = [0.025, 0.03, 0.035, 0.04, 0.045, 0.05, 0.055, 0.06, 0.065, 0.07, 0.075, 0.08]
result_dict1={}
l = len(epsilon_list)
for i in range(l):
    tp = tp_fp_list[i][0]
    fp = tp_fp_list[i][1]
    result_dict1[epsilon_list[i]]={'TP':tp,'FP':fp,'TN':720-14-fp,'FN':14-tp}
TPR_list1 = []
FPR_list1 = []
FNR_list1 = []
TNR_list1 = []
for i in range(l):
    tp = result_dict1[epsilon_list[i]]['TP']
    fn = result_dict1[epsilon_list[i]]['FN']
    fp = result_dict1[epsilon_list[i]]['FP']
    tn = result_dict1[epsilon_list[i]]['TN']
    TPR_list1.append(TPR(tp, fn))
    FPR_list1.append(FPR(fp, tn))
    FNR_list1.append(FNR(fn, tp))
    TNR_list1.append(TNR(tn, fp))
```

```
fig = plt.figure()
plt.clf()
plt.xlabel('Epsilon')
plt.ylabel('FNR')
plt.plot(epsilon_list, FNR_list1,'b', label = 'FNR')
plt.title('Run_Number=100, Same Set of Hash Functions')
plt.scatter(epsilon_list, FNR_list1,c='r')
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['boxplot.showbox'] = True
mpl.rcParams['axes.titlesize']=24
mpl.rcParams['axes.labelsize']=20
mpl.rcParams['lines.linewidth']=2
mpl.rcParams['xtick.labelsize']=10
mpl.rcParams['ytick.labelsize']=10
#mpl.rcParams.keys()
plt.style.use(['grayscale'])
plt.legend()
```
