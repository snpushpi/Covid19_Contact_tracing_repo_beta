```
def TPR(tp, fn):
    frac = tp/(tp+fn)
    return 100*frac
```

```
def FPR(fp, tn):
    frac = fp/(fp+tn)
    return 100*frac
```

```
def tpr_list_generator(group_list):
    result_list = []
    for elt in group_list:
        temp_list = []
        for e in elt:
            temp_list.append(TPR(e,14-e))
        result_list.append(temp_list)
    return result_list
        
```

```
def fpr_list_generator(group_list):
    result_list = []
    for elt in group_list:
        temp_list = []
        for e in elt:
            temp_list.append(FPR(e,720-14-e))
        result_list.append(temp_list)
    return result_list
        
```
